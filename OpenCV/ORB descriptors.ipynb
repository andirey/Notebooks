```python
import cv2 
print(cv2.__version__)
```

```python
plt.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
img = cv2.imread('i3.jpg')
```

```python
plt.imshow(img)
```

```python
orb=cv2.ORB_create()
```

```python
kp = orb.detect(img, None)
```

```python
type(kp)
```

```python
kp, des = orb.compute(img, kp)
```

```python
type(des)
```
