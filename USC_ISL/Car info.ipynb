```python
import pandas as pd
```

```python
cars = pd.read_csv('http://www-bcf.usc.edu/~gareth/ISL/Auto.csv')
```

```python
print(cars.head())
```

```python
print(cars.describe())
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.scatter(cars.weight, cars.mpg)
plt.xlabel('Weight')
plt.ylabel('MPG')
```

```python
plt.scatter(cars.weight, cars.acceleration)
plt.xlabel('Weight')
plt.ylabel('Acceleration')
```

```python
plt.scatter(cars.mpg, cars.displacement)
plt.xlabel('MPG')
plt.ylabel('Displacement')
```
